## 2.Bell States Types?

Let's first understand Quantum entanglement : Quantum entanglement is a phenomenon where two or more qubits become correlated in such a way that the quantum state of each qubit cannot be described independently.

Bell states,are specific quantum states of two qubits that are maximally entangled. There are ùó≥ùóºùòÇùóø Bell states, which are crucial in quantum computing for demonstrating quantum entanglement.

* Each of these quantum pills represents a different type of Bell state, a fundamental concept in quantum computing that showcases the mysterious phenomenon of quantum entanglement. Whether ùóîùóπùó∂ùó∞ùó≤ ùóÆùóªùó± ùóïùóºùóØ are perfectly in sync or always mirroring each other, they are forever connected‚Äîjust like the two halves of these magical pills!

### Breakdown of Bell States


image.png

Let‚Äôs break down the four Bell states using a pill analogy to make it even more relatable and fun.

1. Œ¶+ (ùó£ùóµùó∂-ùóΩùóπùòÇùòÄ) ùòÄùòÅùóÆùòÅùó≤: |Œ¶+‚ü© = 1/‚àö2 (|00‚ü© + |11‚ü©)
This state is an equal superposition of both qubits being in the |0‚ü© state and both qubits being in the |1‚ü© state. If you measure either qubit, you have a 50% chance of getting 0 and a 50% chance of getting 1. However, whatever result you get, you know the other qubit will have the same value.

* ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ: The "ùó£ùó≤ùóøùó≥ùó≤ùó∞ùòÅ ùó†ùóÆùòÅùó∞ùóµ" Pill - It's like Alice and Bob each taking half of a mood-stabilizing pill. If Alice feels calm (blue), Bob also feels calm. If Alice feels energetic (yellow), Bob also feels energetic. Their moods always match, no matter the distance between them.

2. Œ¶- (ùó£ùóµùó∂-ùó∫ùó∂ùóªùòÇùòÄ) ùòÄùòÅùóÆùòÅùó≤:|Œ¶-‚ü© = 1/‚àö2 (|00‚ü© - |11‚ü©)
This is similar to Œ®+, but with a negative sign before the |11‚ü© term. This negative sign introduces a phase difference between the two terms, which is crucial in quantum mechanics. Like Œ®+, measuring either qubit gives a 50% chance of 0 or 1, and the results will always match.

* ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ: The "ùó§ùòÇùóÆùóªùòÅùòÇùó∫ ùóßùòÑùó∂ùóªùòÄ" Pill-Imagine a pill that makes Alice and Bob experience the same mood, but they arrive at it differently. If they both end up calm (blue), Alice might have gradually relaxed while Bob suddenly mellowed out. The end result is the same, but the journey differs.

3. Œ®+ (ùó£ùòÄùó∂-ùóΩùóπùòÇùòÄ) ùòÄùòÅùóÆùòÅùó≤:|Œ¶+‚ü© = 1/‚àö2 (|01‚ü© + |10‚ü©)
This state is an equal superposition of the first qubit being 0 and the second 1, and vice versa. If you measure either qubit, you still have a 50% chance of getting 0 or 1. However, whatever result you get, you know the other qubit will have the opposite value.

* ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ: The "ùóñùóºùó∫ùóΩùóπùó≤ùó∫ùó≤ùóªùòÅùóÆùóøùòÜ ùóñùóºùóπùóºùóøùòÄ" Pill - This is like a balance pill where one person's energy is always complemented by the other's calmness. If Alice takes her half and feels energized (yellow), Bob's half makes him feel calm (blue). They're always in perfect balance, one active and one relaxed.

4.  Œ®- (ùó£ùòÄùó∂-ùó∫ùó∂ùóªùòÇùòÄ) ùòÄùòÅùóÆùòÅùó≤: |Œ®-‚ü© = 1/‚àö2 (|01‚ü© - |10‚ü©)
This is similar to Œ®+, but with a negative sign before the |10‚ü© term. Again, this negative sign introduces a phase difference. Like Œ®+, measuring either qubit gives a 50% chance of 0 or 1, and the results will always be opposite.

* ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ: The "ùó§ùòÇùóÆùóªùòÅùòÇùó∫ ùó¢ùóΩùóΩùóºùòÄùó∂ùòÅùó≤ùòÄ" Pill - Think of this as an advanced version of the balance pill. Not only does it make Alice and Bob experience opposite states (energized vs. calm), but it also influences how they transition between states. If Alice's energy gradually increases, Bob might experience a sudden onset of calmness.

### 2. Importing Qiskit and Required Modules
* Here, we import QuantumCircuit for creating our quantum circuit, Aer for using simulators, and the transpile function to run our circuit.


```python
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
```

### 3. Creating a Quantum Circuit

* We will start by creating a quantum circuit with two qubits.

```python
# Step 1: Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)
```

### 4. Applying Quantum Gates to Create a Bell State
#### To create a Bell state, we need to apply two operations:

##### * Hadamard Gate (H-gate): ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ - This puts the first qubit in a superposition.This gate puts the first qubit (Alice) into a superposition of states |0‚ü© and |1‚ü©, like giving her a pill that makes her both awake and asleep at the same time.

##### * qc.h(0) applies a Hadamard gate on qubit 0, creating a superposition state:

##### * CNOT Gate: ùó•ùó≤ùóÆùóπ ùóüùó∂ùó≥ùó≤ ùóîùóªùóÆùóπùóºùó¥ùòÜ - This entangles the second qubit with the first.The second qubit (Bob) is now entangled with Alice‚Äôs state. The pill now affects both Alice and Bob, connecting their states. If Alice is awake (|1‚ü©), Bob will be awake (|1‚ü©) too. If Alice is asleep (|0‚ü©), Bob will be asleep (|0‚ü©) as well.

##### * qc.cx(0, 1) applies a CNOT gate. The first qubit (qubit 0) is the control qubit, and the second qubit (qubit 1) is the target qubit. This gate entangles the two qubits.

```python
# Apply Hadamard gate on the first qubit
qc.h(0)

# Apply CNOT gate with the first qubit as control and second qubit as target
qc.cx(0, 1)
```

### 5. Measuring the Qubits
* To observe the quantum state, we need to measure the qubits.

* This measures qubit 0 and stores the result in classical bit 0, and measures qubit 1 into classical bit 1*

```python
# Measure the qubits
qc.measure([0, 1], [0, 1])
```

### 6.Circuit Simulation

* Simulates the quantum circuit and the measurement results (counts) to ensure that the circuit is correct and returning values.

```python
 # Create a simulator
simulator = AerSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the simulation and get the result
result = simulator.run(compiled_circuit).result()
```

 ### 7. Visualize the Circuit

```python
# Print the circuit
print("Quantum Circuit:\n")
print(qc.draw())
```

### 8. Print the Result

```python
# Print the result
counts = result.get_counts()
print("\nMeasurement Results (Counts):\n", counts)
```

## 6.Interpretation

This result is typical of a Bell state measurement in quantum computing.

Bell states: These are maximally entangled quantum states of two qubits. There are four Bell states, but this result looks like it's from one of two specific Bell states:
|Œ¶+‚ü© = (|00‚ü© + |11‚ü©) / ‚àö2
or
|Œ¶-‚ü© = (|00‚ü© - |11‚ü©) / ‚àö2
Measurement results:

'00': 539 counts
'11': 485 counts


#### ùóòùòÖùóΩùóπùóÆùóªùóÆùòÅùó∂ùóºùóª:

The measurement shows only two outcomes: '00' and '11'. This is consistent with the Bell states |Œ¶+‚ü© or |Œ¶-‚ü©.
In an ideal Bell state, you'd expect exactly 50% '00' and 50% '11' measurements.
The actual results (539 vs 485) are close to a 50-50 split, with some statistical variation.


#### ùó¶ùó∂ùó¥ùóªùó∂ùó≥ùó∂ùó∞ùóÆùóªùó∞ùó≤:

This result strongly suggests that the qubits were indeed in a Bell state.
The slight imbalance between '00' and '11' could be due to:
a) Normal statistical fluctuations
b) Small imperfections in the quantum circuit or measurement process
c) Decoherence or noise in the quantum system
